# ODE

## Ejercicio 1
$y'= -5y+5t^2+2t$
* $0 \le t \le 1$
* $y(t_0)=1/3$
* $N=10$



```python
%autoreload 2
from src import ODE_euler
import math
import numpy as np
```

## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



```python
def f(t, y):
    return -5*y + 5*t**2 + 2*t
 
a = 0
y0 = 1/3
b = 1
N = 10

y_values, t_values, h = ODE_euler(a=a, b=b, f=f, y_t0=y0, N=N)

print("t_values")
for t in t_values:
    print(f"{t:.5f}")

print("\ny_values")
for y in y_values:
    print(f"{y:.5f}")

def real(t):
     return t**2 + (1/3) * np.exp(-5 * t)

t_real = np.linspace(a, b, 100)
y_real = real(t_real)
```

```python
from matplotlib import pyplot as plt

plt.plot(t_values, y_values, label='Euler con N=10', marker='o',linestyle='--')
plt.plot(t_real, y_real, label=' Solucion Exacta')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solución ODE usando el método de Euler')
plt.legend()
plt.grid()
plt.show()
```

# Errores
Calcule el error relativo para diferentes valores de N.

```python
# Calcular error relativo para 10
errores_relativos = [abs((y_r - y_e) / y_r) for y_r, y_e in zip(y_real, y_values[:len(y_real)])]

print("Errores relativos:")
for i, error in enumerate(errores_relativos):
    print(f"t = {t_values[i]:.2f}, error relativo = {error:.5f}")

# Calcular el promedio del error relativo
promedio_error_relativo = sum(errores_relativos) / len(errores_relativos)
print(f"Promedio del error relativo = {promedio_error_relativo:.5f}")
```
